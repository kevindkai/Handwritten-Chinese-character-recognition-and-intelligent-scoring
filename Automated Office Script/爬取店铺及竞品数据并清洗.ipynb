## 店铺清洗

```python
import os 
import pandas as pd
import numpy as np
def data_clear(path):
    data = pd.read_csv(path,encoding="utf8",header=None, delimiter=',')
    try:
        data = data.dropna()
        data.index = data.reset_index().index
    except:
        pass
    columns=["产品名称","单价(元)","交易人数","店铺名称","店铺地址","采集日期"]
    data.columns = columns
    # 清洗单价
    for i in range(len(data["单价(元)"])):
        try:
            data["单价(元)"][i]=data["单价(元)"][i].split("元")[0]
        except:
            pass
    # 清洗交易人数    
    for i in range(len(data["交易人数"])):
        try:
            data["交易人数"][i]=data["交易人数"][i].split("人")[0]
        except:
            pass
        try:
            data["交易人数"][i]=data["交易人数"][i].split("+")[0]
            if "万" in data["交易人数"][i]:
                data["交易人数"][i]=data["交易人数"][i].split("万")[0]
                data["交易人数"][i]=float(data["交易人数"][i])*10000
            else:
                pass
        except:
            pass
    data["销售额"]=0
    na = path.split("/")[-1]
    na = na.split(".")[0]
    data["品牌名"]= na
    data["采集人员"]="gz"
    data["产品类别"]=0
    data["店铺类别"]=0
    data["省"]=0
    data["市"]=0
    
    # 计算销售额
    for i in range(len(data)):
        data["销售额"][i]=float(data["单价(元)"][i])*float(data["交易人数"][i])
        try:
            data["省"][i]=data["店铺地址"][i].split(" ")[0]
            data["市"][i]=data["店铺地址"][i].split(" ")[1]
        except:
            data["省"][i]=data["店铺地址"][i]
            data["市"][i]=data["店铺地址"][i]
            
        if "旗舰店" in data["店铺名称"][i]:
            data["品牌名"][i] = data["店铺名称"][i].split("旗舰店")[0]
            data["店铺类别"][i]= "旗舰店"
        elif "专卖店" in data["店铺名称"][i]:
            data["品牌名"][i] = data["店铺名称"][i].split("专卖店")[0]
            data["店铺类别"][i]= "专卖店"
        elif "专营店" in data["店铺名称"][i]:
            data["品牌名"][i] = data["店铺名称"][i].split("专营店")[0]
            data["店铺类别"][i]= "专营店"
        elif "工厂店" in data["店铺名称"][i]:
            data["品牌名"][i] = data["店铺名称"][i]
            data["店铺类别"][i]= "工厂店"
        elif "文具店" in data["店铺名称"][i]:
            data["品牌名"][i] = data["店铺名称"][i].split("文具店")[0]
            data["店铺类别"][i]= "文具店"
        elif "河小象" in data["店铺名称"][i]:
            data["品牌名"][i] = "河小象"
            data["店铺类别"][i]= "河小象"
        else:
            data["品牌名"][i]="其它"
            data["店铺类别"][i]= "其它"
        
    for i in range(len(data)):
        if "凹槽" in data["产品名称"][i]:
            data["产品类别"][i]="凹槽"
        elif "水写布" in data["产品名称"][i]:
                data["产品类别"][i]= "水写布"
        elif "笔架" in data["产品名称"][i]: 
            data["产品类别"][i]="笔架"
        elif "钢笔纸" in data["产品名称"][i]: 
            data["产品类别"][i]="钢笔纸"
        elif "簪花小楷" in data["产品名称"][i]: 
            data["产品类别"][i]="簪花小楷"
        elif "佛经字帖" in data["产品名称"][i]: 
            data["产品类别"][i]="佛经字帖"
        elif "国画工具套装" in data["产品名称"][i]: 
            data["产品类别"][i]="国画工具套装"
        elif "文房四宝套装" in data["产品名称"][i]: 
            data["产品类别"][i]="文房四宝套装"
        elif "毛边纸" in data["产品名称"][i]: 
            data["产品类别"][i]="毛边纸"
        elif "宣纸" in data["产品名称"][i]: 
            data["产品类别"][i]="宣纸"
        elif "毛笔字帖" in data["产品名称"][i]: 
            data["产品类别"][i]="毛笔字帖"
        else:
            data["产品类别"][i]=0
        
    # 删除重复数据    
    data = data.drop_duplicates()
    # 删除销售额为0的数据 
    data = data[data["销售额"]!=0]        
    data.to_csv("clear/clear_"+na+".csv",encoding="utf-8-sig",index=None)
```

## 添加时间及类目

```python
data = pd.read_csv("生e经.csv",encoding="utf-8")
data.head()
```

```python
data["一级类目"]=0
data["二级类目"]=0
data["三级类目"]=0
for i in range(len(data)):
    try:
        df = data["行业类目"][i].split("->")
        if len(df)==3:
            data["一级类目"][i]=df[0]
            data["二级类目"][i]=df[1]
            data["三级类目"][i]=df[2]
        elif len(df)==2:
            data["一级类目"][i]=df[0]
            data["二级类目"][i]=df[1]
            data["三级类目"][i]=df[1]
        else:
            data["一级类目"][i]=df[0]
            data["二级类目"][i]=df[0]
            data["三级类目"][i]=df[0]
    except:
        pass
```

## 更改列的顺序

```python
cols = list(data)
cols
cols.insert(0,cols.pop(cols.index('日期')))
cols.insert(5,cols.pop(cols.index('一级类目')))
cols.insert(6,cols.pop(cols.index('二级类目')))
cols.insert(7,cols.pop(cols.index('三级类目')))
data = data.loc[:,cols]
```
