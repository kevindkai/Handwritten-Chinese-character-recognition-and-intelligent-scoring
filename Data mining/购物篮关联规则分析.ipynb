```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"E:\研究生资料\47448_R语言商务数据分析实战_源代码和实验数据\R语言商务数据分析实战\第2章\任务程序\data\GoodsOrder.csv")
data.columns=["ID","Goods"]
data.head()
```

```python
freq=data['Goods'].value_counts(normalize=False)
prob=data['Goods'].value_counts(normalize=True)
prob=pd.DataFrame(prob)
freq=pd.DataFrame(freq)
newdata=pd.merge(freq,prob,on=freq.index)
newdata.columns=["商品类别","销量","销量占比"]
newdata.head()
```

```python
d=data["ID"].value_counts()
d.index
```

```python
result=[]
for index in (d.index):
    x=np.array(data[data.iloc[:,0]==index].iloc[:,1])
    x=x.tolist()
    result.append(x)
```

```python
te=TransactionEncoder()
```

```python
te_ary=te.fit_transform(result)
df=pd.DataFrame(te_ary,columns=te.columns_)
freq=apriori(df,min_support=0.05,use_colnames=True)
```

```python
df
```

```python
from mlxtend.frequent_patterns import association_rules
final_result=association_rules(freq,metric='confidence',min_threshold=0.1)
final_result
```
