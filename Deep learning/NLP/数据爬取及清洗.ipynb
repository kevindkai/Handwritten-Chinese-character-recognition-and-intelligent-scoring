```python
def get_urls(http):
    response=requests.get(http,headers=headers)
    text = response.text
    html=etree.HTML(text)
    detail_urls=html.xpath("//h3/a/@href")
    urls=[]
    for detail_url in detail_urls:
        url="https://car.autohome.com.cn/"+detail_url
        urls.append(url)
    return urls
```

```python
def detail_parse(url):
    response=requests.get(url,headers=headers)
    text = response.text
    html=etree.HTML(text)
    id=html.xpath("//dl/dd/a/@id")
    return id
```

```python
def get_contents(car_url):
    response=requests.get(car_url,headers=headers)
    text = response.text
    html=etree.HTML(text)
    car_name=html.xpath("//span[@class='car-name fn-left']/a/text()")
    title=html.xpath("//div[@class='title-name name-width-01']/a/text()")
    content=html.xpath("//div[@style='font-family: myfont,宋体,arial,tahoma,sans-serif;']/text()")
    return car_name,title,content
```
